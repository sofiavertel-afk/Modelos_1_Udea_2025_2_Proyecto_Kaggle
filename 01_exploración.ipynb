```python
# üß† Proyecto UdeA 2025-2
# Exploraci√≥n inicial de h√°bitos digitales y rendimiento acad√©mico

# 1Ô∏è‚É£ Importar librer√≠as b√°sicas
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# 2Ô∏è‚É£ Crear un dataset de ejemplo (simulando "train.csv" de Kaggle)
import numpy as np
import pandas as pd

np.random.seed(42)
data = pd.DataFrame({
    "horas_estudio": np.random.normal(3, 1.2, 100),
    "horas_sueno": np.random.normal(7, 1.5, 100),
    "uso_celular_horas": np.random.normal(4, 1.3, 100),
    "rendimiento": np.random.choice(["Alto", "Medio", "Bajo"], 100)
})

data.head()
```

```python
# 3Ô∏è‚É£ Mostrar las primeras filas del dataset
print("üìã Vista inicial de los datos:")
display(data.head())

# 4Ô∏è‚É£ Informaci√≥n general del conjunto de datos
print("\n‚ÑπÔ∏è Informaci√≥n general:")
print(data.info())
```

```python
# 5Ô∏è‚É£ Estad√≠sticas descriptivas
print("\nüìä Estad√≠sticas descriptivas:")
display(data.describe())

# 6Ô∏è‚É£ Visualizaci√≥n: distribuci√≥n de horas de estudio
if "horas_estudio" in data.columns:
    plt.figure(figsize=(6,4))
    sns.histplot(data["horas_estudio"], bins=10, kde=True)
    plt.title("Distribuci√≥n de horas de estudio")
    plt.xlabel("Horas de estudio")
    plt.ylabel("Frecuencia")
    plt.show()
else:
    print("‚ö†Ô∏è No existe la columna 'horas_estudio' en el dataset.")
```

```python
# 7Ô∏è‚É£ Relaci√≥n entre uso del celular y rendimiento
if all(col in data.columns for col in ["rendimiento", "uso_celular_horas"]):
    plt.figure(figsize=(6,4))
    sns.boxplot(x="rendimiento", y="uso_celular_horas", data=data, palette="viridis")
    plt.title("Relaci√≥n entre uso del celular y rendimiento acad√©mico")
    plt.show()
else:
    print("‚ö†Ô∏è Faltan columnas: 'rendimiento' o 'uso_celular_horas'.")
```

```python
# 8Ô∏è‚É£ Matriz de correlaci√≥n
cols_corr = [c for c in ["horas_estudio", "horas_sueno", "uso_celular_horas"] if c in data.columns]
if len(cols_corr) >= 2:
    plt.figure(figsize=(5,4))
    sns.heatmap(data[cols_corr].corr(), annot=True, cmap="coolwarm")
    plt.title("Matriz de correlaci√≥n entre variables num√©ricas")
    plt.show()
else:
    print("‚ö†Ô∏è No hay suficientes variables num√©ricas para calcular correlaci√≥n.")
```
